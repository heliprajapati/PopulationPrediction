```python
# Import libraries
import pandas as pd
import numpy as np
import json
import os
import pandas_profiling
from pydantic.v1 import BaseSettings
from pandas_profiling import ProfileReport
```

```python
# Define the directory containing JSON files
directory_path = 'C:\\Users\\helip\\Downloads\\data-pipelines-main\\1'

# Initialize an empty list to hold the data frames
data_frames = []

# Loop through all files in the directory
for filename in os.listdir(directory_path):
    if filename.endswith('.json'):
        file_path = os.path.join(directory_path, filename)
        
        # Load each JSON file
        with open(file_path, 'r') as file:
            data = json.load(file)
        
        # Convert JSON object to data frame and append to the list
        df = pd.json_normalize(data)
        data_frames.append(df)

# Combine all data frames into one
combined_data = pd.concat(data_frames, ignore_index=True)

# Inspect the combined data
combined_data.head(5)

```

```python
# Check the structure of the combined data
print(combined_data.info())
print(combined_data.describe())
```

```python
# Exploratory Data Analysis
profile = ProfileReport(combined_data, title='Data Profiling Report')
profile.to_notebook_iframe()
```
